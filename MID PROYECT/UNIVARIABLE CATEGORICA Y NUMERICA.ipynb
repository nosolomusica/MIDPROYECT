## CATEGORICA NUMERICAS BIVARIBALE

## 1. AnÃ¡lisis de la distribuciÃ³n del tempo por gÃ©nero

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="label", y="tempo", palette="muted")
plt.title("DistribuciÃ³n del Tempo por GÃ©nero", fontsize=16)
plt.xlabel("GÃ©nero", fontsize=12)
plt.ylabel("Tempo (BPM)", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Calcular los valores mÃ¡s altos y mÃ¡s bajos por gÃ©nero
min_max_values = df.groupby('label')['tempo'].agg(['min', 'max']).reset_index()
generos = min_max_values['label']
min_values = min_max_values['min']
max_values = min_max_values['max']
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Dibujar barras para valores mÃ­nimos y mÃ¡ximos
ax.bar(generos, min_values, color='blue', label='MÃ­nimo')
ax.bar(generos, max_values, color='orange', label='MÃ¡ximo', alpha=0.7)

# Etiquetas y tÃ­tulo
ax.set_xlabel('label', fontsize=12)
ax.set_ylabel('tempo', fontsize=12)
ax.set_title('Valor mÃ­nimo y mÃ¡ximo por gÃ©nero', fontsize=14)
ax.legend()

# Mostrar el grÃ¡fico
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##Â 2. AnÃ¡lisis de la distribuciÃ³n de beats por gÃ©nero

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="label", y="beats", palette="muted")
plt.title("DistribuciÃ³n de beats por GÃ©nero", fontsize=16)
plt.xlabel("GÃ©nero", fontsize=12)
plt.ylabel("Beats", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Calcular los valores mÃ¡s altos y mÃ¡s bajos de beats por gÃ©nero 
min_max_values = df.groupby('label')['beats'].agg(['min', 'max']).reset_index()
generos = min_max_values['label']
min_values = min_max_values['min']
max_values = min_max_values['max']
```

```python
# Crear el grÃ¡fico
fig, ax = plt.subplots(figsize=(10, 6))

# Dibujar barras para valores mÃ­nimos y mÃ¡ximos
ax.bar(generos, min_values, color='skyblue', label='MÃ­nimo')
ax.bar(generos, max_values, color='orange', label='MÃ¡ximo', alpha=0.7)

# Etiquetas y tÃ­tulo
ax.set_xlabel('label', fontsize=12)
ax.set_ylabel('beats', fontsize=12)
ax.set_title('Valor mÃ­nimo y mÃ¡ximo de beats por gÃ©nero', fontsize=14)
ax.legend()

# Mostrar el grÃ¡fico
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
